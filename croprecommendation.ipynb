```python
import pandas as pd
```

```python
df=pd.read_csv("/content/Crop_recommendation.csv")
df.head(6)
```

```python
df['label'].value_counts()
```

```python
x=df.drop('label',axis=1)
y=df['label']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, random_state=1, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred1 = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
logistic_reg_acc=accuracy_score(y_test,y_pred1)
print("logistic accuracy is: " + str(logistic_reg_acc))
```

```python
from sklearn.tree import DecisionTreeClassifier
model2=DecisionTreeClassifier()
model2.fit(x_train, y_train)
y_pred2=model2.predict(x_test)
```

```python
decision_acc=accuracy_score(y_test,y_pred2)
print("Decision tree accuracy is: " + str(decision_acc))
```

```python
from sklearn.ensemble import RandomForestClassifier
model3 = RandomForestClassifier()
model3.fit(x_train,y_train)
y_pred3=model3.predict(x_test)
```

```python
randomforest_acc=accuracy_score(y_test,y_pred3)
print("Random Forest accuracy is: " + str(randomforest_acc))
```
